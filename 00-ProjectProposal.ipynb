## Research Question

Is there a higher incidence of anxiety and depression among college-aged adults (18-25) that use social media more frequently?  
We will look at how many hours per day a person uses social media, categorized by platform, and relate that to mental health diagnostics given by certified health care professionals. This will be a statistical inference work, as it will be looking among datasets and exploring how strongly they are linked with each other. 

## Background and Prior Work

The relationship between social media use and mental health has rapidly become one of the most heavily examined topics in psychological research over the past decade. As of now, more than 30% of college students report experiencing depression, and approximately 90% of young adults today report utilizing social media in their daily lives – roughly eight times the proportion that did so in 2005. Due to the speed of which this percentage changed, researchers across the globe have dedicated tremendous effort to further examining whether social media use contributes to rising depression and anxiety levels among college-aged students. 

Young Adult Depression and Anxiety Linked to Social Media Use: Assessment and Treatment” has examined the relationship between social media use and mental health specifically among college-aged adults. Bettmann reviewed existing literature on this topic and found that social media use is consistently associated with depression and anxiety in young adults.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The authors showed that as many as 90% of young adults currently use social media compared to only 12.5% in 2005. It also highlighted the fact that more than 30% of college students are currently depressed.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Significant results of the research were that it is not simply the amount of time spent on social media that matters. However, how you use really mattered. Aspects such as passive scrolling and frequent social comparison were particularly strongly linked to depressive and anxious symptoms. Moreover, young adults who used a larger variety of platforms were significantly more likely to show symptoms of depression and anxiety compared to those who used only few. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The authors also introduced the concept of Fear of Missing Out (FoMO), which is the anxiety that others are having more rewarding experiences, as a key mechanism that links social media usage to depression and anxiety. However, the review also acknowledged that the direction of cause and effect remains unclear. It is questionable whether social media use leads to worse mental health, or whether individuals who already experience depression and anxiety are simply more drawn to heavy social media use. This ambiguity would also be an important factor that our project should consider.

In an attempt to empirically examine how social media use correlates with mental health indicators, Osman conducted a cross-sectional study at Mogadishu University. Out of the 268 participants he interviewed, 84.7% reported spending more than three hours per day using social media.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) High levels of social media usage proved a strong correlation between sleep disturbance, anxiety, and social isolation. Of the three, anxiety outcomes proved to be the highest (AOR: 22.2; 95% CI: 3.73-131.90).<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) These results are particularly noteworthy, as the finding regarding anxiety outcomes showcases that independent screen usage – despite specific content viewed or platform type – may be enough information to predict anxiety levels in a population made up of college-aged students.

On a grander scale, Wang, Zhang, and Xiao (2025) further investigated these associations in a large-scale cross-cultural study of 1,500 Chinese university students drawn from ten universities across China.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Utilizing the Depression Anxiety Stress Scale-21 (DASS-21) and the Social Media Use Integration Scale (SMUIS), the authors found that there were indeed positive associations between social media usage and both depressive and anxiety symptoms. This study also identified sleep quality as a significant mediator within this relationship, as disrupted sleep proved to account for roughly 30.4% of the total effect of social media use on depression. This suggests that disrupted sleep could be a manner through which heavy platform use actively degrades mental health.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Additionally, the association between social media usage and the presence of anxiety and depressive symptoms was found to be higher among female students, a finding that should be treated as a potential confounding variable in further research/analysis done on this topic.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

In totality, each of these studies suggest that there is a consistent correlation between social media use and depression and anxiety symptoms in college-aged individuals. These studies also go out of their way to acknowledge important moderating factors, such as gender and sleep quality, and are valid by utilizing validated screening instruments and rigorous sampling methods. Ultimately, by using validated screening instruments and appropriate sampling methods, these studies provide a strong empirical foundation for our project.
